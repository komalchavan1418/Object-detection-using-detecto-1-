```python
import matplotlib.pyplot as plt
from detecto.utils import read_image
```

```python
base_path="/content/drive/MyDrive/train/"
sample_image="orange_70.jpg"
image=read_image(base_path+sample_image)
plt.imshow(image)
plt.show() ##### for one individual image showing
```

```python
from detecto.core import Dataset
dataset=Dataset(base_path)
```

```python
from detecto.visualize import show_labeled_image
image,targets=dataset[10]
show_labeled_image(image,targets['boxes'],targets['labels'])    #### observing the labeling 
```

```python
from detecto.core import DataLoader,Model           ### model fiting

labels=["apple","banana","mixed","orange"]

model=Model(labels)
```

```python
model.fit(dataset)
```

```python
import cv2
```

```python
image=read_image(base_path+"apple_8.jpg")
labels,boxes,scores=model.predict(image)
```

```python
print("labels",labels)
print("boxes",boxes)
print("scores",scores)
```

```python
from detecto.visualize import show_labeled_image
show_labeled_image(image,boxes[0],labels[0])
```
